## Part 2 : Concatenate a dataframe and Append

```python
append_to_df1 = []

append_to_df1.append(["123", "Omar", "36"])

df1 = pd.DataFrame(append_to_df1, columns = ["ID", "Name", "Age"])
df1.head()
```

```python
append_to_df2 = []
append_to_df2.append(["345", "Anna", "19"])

df2 = pd.DataFrame(append_to_df2, columns=["ID", "Name", "Age"])
df2.head()
```

```python
combined_df = pd.concat([df1, df2], ignore_index = True, sort = False)
combined_df
```
